# Protein Folding using Genatic Algorithm (GA) - Triangular Lattice

This notebook applies the **Genatic Algorithm (GA)** method to protein 
folding in the `HP` model on a **Triangular lattice**. 

We benchmark multiple protein **`HP` sequences** (already converted from FASTA), 
record the **minimum energy** achieved and **runtime**, and save the results in a structured dataset.

```python
import pandas as pd

dataset_hp = pd.read_excel(r"C:\Users\Yassin\OneDrive\Desktop\Work\Machine Learning\Machine Learning Internship\Github-1\Protein_Folding_Model-3\data\HP_dataset.xlsx")
dataset_hp.index = range(1, len(dataset_hp) + 1)
dataset_hp
```

### **Id 1 Sequence**

```python
import time
from genatic_algorithm_method import trianguler_genetic_algorithm, visualize_fold

protein_id = "1"
hp_sequence = "HPHPPHHPHPPHPHHPPHPH"

lattice_type = "triangular"  

start_time = time.time()

if lattice_type == "square":
    min_energy, best_coords = square_genetic_algorithm(hp_sequence, generations=5000, pop_size=200)
elif lattice_type == "triangular":
    min_energy, best_coords = trianguler_genetic_algorithm(hp_sequence, generations=5000, pop_size=500)
elif lattice_type == "cubic":
    min_energy, best_coords = genetic_algorithm_3d(hp_sequence, generations=5000, pop_size=200)
else:
    raise ValueError("Invalid lattice type selected.")

end_time = time.time()

runtime = round((end_time - start_time) / 60, 2)

# Print results
print(f"Protein Id: {protein_id}")
print(f"HP Sequence: {hp_sequence}")
print(f"Length: {len(hp_sequence)}")
print(f"Minimum Energy: {min_energy}")
print(f"Runtime (min): {runtime}")

# Visualize folding
visualize_fold(hp_sequence, best_coords, min_energy)
```
