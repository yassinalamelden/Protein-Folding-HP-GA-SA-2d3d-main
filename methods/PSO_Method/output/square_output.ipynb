# Protein Folding using Particle Swarm Optimization (PSO) - Square Lattice

This notebook applies the **Particle Swarm Optimization (PSO)** method to protein 
folding in the `HP` model on a **Square lattice**. 

We benchmark multiple protein **HP sequences** (already converted from FASTA), 
record the **minimum energy** achieved and **runtime**, and save the results in a structured dataset.


```python
import pandas as pd
import time
from pso_method import square_pso
```

```python
dataset_hp = pd.read_excel("HP_dataset.xlsx")
dataset_hp.index = range(1, len(dataset_hp) + 1)
dataset_hp
```

### **Id 1 Sequence**

```python
import time
from pso_method import square_pso, visualize_fold

protein_name = "1"
hp_sequence = "HPHPPHHPHPPHPHHPPHPH"

start_time = time.time()
min_energy, best_coords = square_pso(hp_sequence, swarm_size=150, max_iter=5000)
end_time = time.time()

runtime = round((end_time - start_time) / 60, 2)

# Print results
print(f"Protein: {protein_name}")
print(f"HP Sequence: {hp_sequence}")
print(f"Length: {len(hp_sequence)}")
print(f"Minimum Energy: {min_energy}")
print(f"Runtime (min): {runtime}")

# Visualize folding
visualize_fold(hp_sequence, best_coords, min_energy)
```
